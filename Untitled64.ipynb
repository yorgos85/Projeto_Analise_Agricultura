```python
# Importando as Bibliotecas
import pandas as pd
import numpy as np

# carregando csv

df = pd.read_csv('/content/drive/MyDrive/ARQUIVOS ESTUDOS CSV/materia_agriculturta.csv')
```

```python
#Tratando os dados
df = df.replace('%', '', regex = True)
df = df.replace(',', '', regex = True)
df = df.replace('-', '', regex = True)
df = df.replace('', np.nan)
df = df.replace('MAY90', np.nan)

#Apagando valores nulos
df = df.dropna()

#Checa novamente valores nulos novamente
df.isnull().sum()


#colocando os dados em uma lista transformando em FLOAT
conve = ['Coarse wool Price',
         'Coarse wool price % Change',
         'Copra Price',
         'Copra price % Change',
         'Cotton price % Change',
         'Fine wool Price',
         'Fine wool price % Change',
         'Hard log price % Change',
         'Hard sawnwood price % Change',
         'Hide price % change',
         'Plywood price % Change',
         'Rubber price % Change',
         'Softlog price % Change',
         'Soft sawnwood price % Change',
         'Wood pulp price % Change'

]

#CONVERTENDO PARA FLOAT

df [conve] = df[conve].astype('float')

#Trazendo a coluna Month para formato data
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Month')

```

```python
#Iniciando a An√°lise exploratoria
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

%matplotlib inline
sns.set_style('darkgrid')

matplotlib.rcParams['font.size']= 14
matplotlib.rcParams['figure.figsize']= (9,5)
matplotlib.rcParams['figure.facecolor']= 'white'


#lista de precos de produto
raw_data = ['Coarse wool Price',
            'Copra Price',
            'Cotton Price',
            'Fine wool Price',
            'Hard log Price',
            'Hard sawnwood Price',
            'Hide Price',
            'Plywood Price',
            'Rubber Price',
            'Soft sawnwood Price',
            'Wood pulp Price'
]

#ANALISANDO A CORRELACAO DOS PRODUTOS
corrmat = df[raw_data].corr()
#tamanho do plot
fig = plt.figure(figsize=(12,9))

mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8, mask=mask, square= True, annot= True)
plt.show()
```

- Conforme grafico acima podemos concluir que as materiais primas tem poucas relacoes entre uma e outra

```python
#ANALISANDO A CORRELACAO DE VARIACAO DE PRECO

fig = plt.figure(figsize=(30,15))

listchange = ['Copra price % Change',
              'Cotton price % Change',
              'Plywood price % Change', 
               'Fine wool price % Change',
              'Hard log price % Change', 
              'Hard sawnwood price % Change',
              'Hide price % change',
              'Rubber price % Change',
              'Softlog price % Change', 
              'Soft sawnwood price % Change',
              'Wood pulp price % Change',
              'Coarse wool price % Change'
              ]

#Extraindo a correlacao forte acima de 0.7

corrMatriz = df[listchange].corr()
sns.heatmap(corrMatriz, annot= True)
plt.show()
```

- Analisando o grafico podemos concluir que nao correlao forte entre a variacao do preco, uma correlacao forte tem acima de 0,7

```python
#DESCOBRINDO A VARIACAO NORMAL DO PRECO DE CADA MATERIA-PRIMA

listchange = ['Copra price % Change',
              'Cotton price % Change',
              'Plywood price % Change', 
               'Fine wool price % Change',
              'Hard log price % Change', 
              'Hard sawnwood price % Change',
              'Hide price % change',
              'Rubber price % Change',
              'Softlog price % Change', 
              'Soft sawnwood price % Change',
              'Wood pulp price % Change',
              'Coarse wool price % Change'
              ]
for i in range(len(listchange)):
  plt.figure(figsize=(12,12))
  df[listchange[i]].hist(figsize=(11,9), linewidth=1)
  plt.xlabel('% Change')
  plt.ylabel('count')
  plt.legend(listchange[i:], loc= 'upper center', bbox_to_anchor=(1.2,1))
  
```
